What is tuple in python

A tuple is a collection of items which is ordered and unchangeable. It allows duplicate members. Tuples are written with round brackets.
Tuples can contain items of different data types, but usually the items all have the same type.

some methods and attribute in tuple
1. count() - Returns the number of times a specified value occurs in a tuple.
2. index() - Searches the tuple for a specified value and returns the position of where it was found.


How tuple is different from list
1. Syntax: Tuples are defined using parentheses (), while lists are defined using square brackets [].
2. Mutability: Tuples are immutable, meaning their elements cannot be changed after they are created. Lists are mutable, allowing you to modify, add, or remove elements.
3. Performance: Tuples are generally faster than lists because they are immutable and have a smaller memory footprint. Lists require more memory and processing time due to their mutability.
4. Use Cases: Tuples are often used for fixed collections of items, such as coordinates or database records, where immutability is desired. Lists are more suitable for collections that may need to be modified, such as a list of user inputs or a collection of items that may change over time.
