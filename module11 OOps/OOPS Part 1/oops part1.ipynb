```python
## create a class named pwskills and create a method named welcome_msg which will print welcome to pwskills
class pwskills :
    def welcom_msg(self):
        print("Welcome to pwskills")
```

```python
## create an object of the class pwskills and call the method welcome_msg
rohan = pwskills()
```

```python
## to check the type of the object
print(type(rohan))
```

```python
## call the method welcome_msg using the object rohan
rohan.welcom_msg()
```

```python
## create a class named pwskills1 and create a method named return_student_details which will return the student details like phone number, email id and student id
class pwskills1 :
    def __init__(self,phone_number,email_id,student_id):
        self.phone_number = phone_number
        self.email_id=email_id
        self.student_id=student_id

    def return_student_details(self):
        return self.student_id , self.phone_number,self.email_id
```

```python
## create an object of the class pwskills1 and call the method return_student_details
rohan = pwskills1(8507543581, "raoneface@gmail.com", 10)
```

```python
## to check the type of the object
rohan.return_student_details()
```
