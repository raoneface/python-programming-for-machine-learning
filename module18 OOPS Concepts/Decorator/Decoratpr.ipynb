what is decorator in python?

In Python, a decorator is a design pattern that allows you to modify the behavior of a function or class. It is a higher-order function that takes another function as an argument and extends its behavior without explicitly modifying it. Decorators are often used for logging, access control, memoization, and other cross-cutting concerns.      



```python
def test():
    print("this is the start of my fun")
    print("this is my fun to start")
    print(4+5)
    print("this is the end of my fun")
```

```python
test()
```

```python
def decofun(func):
    def inner_dec():
        print("this is start my function")
        func()
        print("This is end of function")
    return inner_dec
```

```python
def test1():
    print(6+7)
```

```python
@decofun
def test1():
    print(6+7)
```

```python
test1()
```

```python
import time
def decorator1(func):
    def timer_inner_deco():
        start_time = time.time()
        func()
        end_time = time.time()
        print(end_time-start_time)
    return timer_inner_deco
```

```python
@decorator1
def test2():
    print(45+36)
```

```python
test2()
```

```python
@decorator1
def test3():
    for i in range(100000000):
        pass
```

```python
test3()
```

```python
def deco1(func):
    def inner_deco():
        print("india")
        func()
        print("bihar")
    return inner_deco
```

```python
## write a decorator funxtion
@deco1
def printMyaddress():
    print("my sweet home")
```

```python
printMyaddress()
```
