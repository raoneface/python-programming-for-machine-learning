what is class method in python?

In Python, a class method is a method that is bound to the class and not the instance of the class. It can be called on the class itself rather than on an instance of the class. Class methods are defined using the @classmethod decorator and take the class as their first argument, conventionally named 'cls'. They can be used to create factory methods, which are methods that return an instance of the class, or to modify class-level attributes.



```python
## first creting a class pwskills
## constructor

class pwskills :

    def __init__(self, name , email) :
        self.email = email
        self.name = name

    ## class method
    def student_detail(self) :
        print(self.name , self.email)
```

```python
## creating object pw
pw  = pwskills("mohan" , "mohan@gmail.com")
```

```python
pw.name
```

```python
pw.email
```

```python
## accesing class method by object
pw.student_detail()
```

## using class method we will overload __init__ function

```python
class pwskills1 :

    def __init__(self, name , email) :
        self.email = email
        self.name = name

    ## class method
    @classmethod
    def details(cls , name , email) :
        return cls(name , email)

    def student_detail(self) :
        print(self.name , self.email)
```

```python
## we can access class method by class name without creating object
pwskills1.details("sudh" ,"sudh@gmail.com")
```

```python
## 
pw1 = pwskills1.details("sudh" ,"sudh@gmail.com")
```

```python
## 
pw1.name
```

```python
pw1.email
```

```python
## we can use student_detail method withou creating object 
pw1.student_detail()
```

```python
pw1.student_detail()
```

```python
class pwskills2 :

    mobile_num = 8507

    def __init__(self, name , email) :
        self.email = email
        self.name = name

    @classmethod
    def change_number(cls,mobile_num1) :
        pwskills2.mobile_num =  mobile_num1

    @classmethod
    def details(cls , name , email) :
        return cls(name , email)

    def student_detail(self) :
        print(self.name , self.email)
```

```python
##  we can access class variable by class name without creating object
pwskills2.mobile_num
```

```python
##  we can change the value of class variable by using class method without creating object
pwskills2.change_number(4567)
```

```python
## 
pwskills2.mobile_num
```

```python
## mobile_num is also acces by student_detail 
class pwskills3 :

    mobile_num = 8507

    def __init__(self, name , email) :
        self.email = email
        self.name = name

    @classmethod
    def change_number(cls,mobile_num1) :
        pwskills3.mobile_num =  mobile_num1

    @classmethod
    def details(cls , name , email) :
        return cls(name , email)

    def student_detail(self) :
        print(self.name , self.email,pwskills3.mobile_num)
```

```python
## one way using mobile_num is also acces by student_detail through object 
pw_obj = pwskills3("sudh","sudh@gmail.com")
pw_obj.student_detail()
```

```python
## another way using mobile_num is also acces by student_detail through class method without creating object
pw = pwskills3.details("shub","shubh@gmail.com")
pw.student_detail()
```


## we can also add externam method to the class 

```python
## mobile_num is also acces by student_detail 
class pwskills4 :

    mobile_num = 8507

    def __init__(self, name , email) :
        self.email = email
        self.name = name

    @classmethod
    def change_number(cls,mobile_num1) :
        pwskills4.mobile_num =  mobile_num1

    @classmethod
    def details(cls , name , email) :
        return cls(name , email)

    def student_detail(self) :
        print(self.name , self.email,pwskills4.mobile_num)
```

```python
def course_detail(cls,course_name) :
    print("course name is ", course_name)
```

```python
pwskills4.course_detail = classmethod(course_detail)
```

```python
pwskills4.course_detail("data science master class")
```
