```python
## for extracting data from range
for i in range(10):
    print(i)
```

```python
## yield is used to generate the value without cunsumption
def test_fib(n):
    a,b =0,1
    for i in range(n):
        yield a
        a,b = b , a + b
```

```python
test_fib(10)
```

```python
for i in test_fib(10):
    print(i)
```

```python
def test_fib1():
    a,b =0,1
    while True:
        yield a
        a,b = b , a + b
```

```python
fib = test_fib1()
```

```python

for i in range(10):
    print(next(fib))
```
