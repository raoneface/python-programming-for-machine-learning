## Buffered Read And write 
- When we read or write a file, the data is not directly read from or written to the file. Instead, it is first stored in a buffer, which is a temporary storage area in memory. The buffer allows for efficient reading and writing of data, as it reduces the number of times the file needs to be accessed.
- When we read a file, the data is read into the buffer in chunks, and we
can access the data from the buffer until it is empty. When we write to a file, the data is first written to the buffer, and then it is flushed to the file when the buffer is full or when we explicitly flush it.
- The buffer can be useful for improving performance, especially when working with large files, as it
reduces the number of disk accesses. However, it is important to be aware of the buffer when working with files, as it can lead to unexpected behavior if not handled properly. For example, if we write to a file and then immediately read from it without flushing the buffer, we may not see the changes we just made.


```python
import io
```

```python
with open("test1.txt" ,"wb") as f :
    file = io.BufferedWriter(f)

    file.write(b"this is my first line\n")
    file.write(b"this is my second line\n")
    file.flush()
```

```python
with open("test1.txt","rb") as f :
    file = io.BufferedReader(f)
    data = file.read()
```

```python
print(data)
```
