```python
data = {
    "name" : "shubham",
    "mail_id" : "raoneface@gmail.com",
    "phone_number" : 8507543581,
    "subject" : ["data science" ,"big data" ,"data analytics"]
}
```

```python
type(data)
```

```python
import json
```

```python
## writing data to json file    
## we can use dump() method to write data to json file
## we have to open the file in write mode and then we can use dump() method to write data to json file
## json.dump(data,f) method takes two arguments first is the data which we want to write to json file and second is the file object which we have opened in write mode
## w is write mode

with open("data.json" ,'w') as f:
    json.dump(data,f)
```

```python
## reading data from json file
## we can use load() method to read data from json file
## we have to open the file in read mode and then we can use load() method to read data from json file
## json.load(f) method takes one argument which is the file object which we have opened in read mode    

with open("data.json",'r') as f :
    data1 = json.load(f)
```

```python
data1
```

```python
## we can access the data from json file using the keys of the dictionary
data1["subject"][1]
```

```python
## working with csv files comma separated values
import csv
```

```python
data =[["name","email_id","phone_number"],["shubham","shubham@gmail.com","1234567890"],["rohit","rohit@gmail.com","0987654321"],["sachin","sachin@gmail.com","1122334455"]]
```

```python
with open("data.csv",'w',newline='') as f:
    writer = csv.writer(f)

    for i in data:
        writer.writerow(i)
```

```python
with open("data.csv",'r') as f:
    read_data = csv.reader(f)

    for i in read_data:
        print(i)
```
