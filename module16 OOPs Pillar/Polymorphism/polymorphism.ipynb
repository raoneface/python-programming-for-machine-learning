What is polymorphism in python?
Polymorphism is a fundamental concept in object-oriented programming (OOP) that allows objects of different classes to be treated as objects of a common superclass. It enables a single interface to represent different underlying data types or classes. In Python, polymorphism can be achieved through method overriding and duck typing.

in easy language
Polymorphism in Python means that you can use the same method or function name to perform different tasks based on the context. It allows you to write code that can work with objects of different types, as long as they have the same method or function name. This makes your code more flexible and easier to maintain. For example, you can have a method called "speak" that behaves differently for a dog and a cat, but you can call it in the same way for both animals.

There are two main types of polymorphism in Python:
1. **Compile-time Polymorphism (Method Overloading)**: This type of polymorphism is achieved by defining multiple methods with the same name but different parameters within the same class. However, Python does not support method overloading in the traditional sense, but you can achieve similar functionality using default arguments or variable-length arguments.



compile time polymorphism in python
As mentioned earlier, Python does not support traditional method overloading (compile-time polymorphism) like some other programming languages. However, you can achieve similar functionality using default arguments or variable-length arguments. Hereâ€™s an example of how to mimic compile-time polymorphism in Python:

```python
class Calculator:
    def add(self, a, b):
        return a + b
    def add(self, a, b, c=0):
        return a + b + c
calc = Calculator()
# Using the add method with two arguments
print(calc.add(5, 10))  # Output: 15
# Using the add method with three arguments
print(calc.add(5, 10, 15))  # Output: 30
```

In this example, we have defined two `add` methods in the `Calculator` class. However, since Python does not support method overloading, the second definition of `add` will overwrite the first one. To achieve similar functionality, we can use default arguments to allow for a variable number of parameters. In this case, the `c` parameter has a default value of 0, allowing us to call the `add` method with either two or three arguments.


2. **Run-time Polymorphism (Method Overriding)**: This type of polymorphism occurs when a subclass provides a specific implementation of a method that is already defined in its superclass. The method

in the subclass overrides the method in the superclass, allowing for different behavior when the method is called on an instance of the subclass.
Here is an example of run-time polymorphism (method overriding) in Python:

```python
class Animal:
    def speak(self):
        return "Animal speaks"
class Dog(Animal):
    def speak(self):
        return "Dog barks"
class Cat(Animal):
    def speak(self):
        return "Cat meows"
# Create instances of Dog and Cat
dog = Dog()
cat = Cat()
# Call the speak method on both instances
print(dog.speak())  # Output: Dog barks
print(cat.speak())  # Output: Cat meows
```
In this example, the `speak` method is defined in the `Animal` class and overridden in both the `Dog` and `Cat` subclasses. When we call the `speak` method on instances of `Dog` and `Cat`, it executes the overridden method specific to each subclass, demonstrating run-time polymorphism.
